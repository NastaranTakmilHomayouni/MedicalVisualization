<b> Visualize the flow with using arrow glyphs ans streamlines </b>

```python
x = np.arange(0,2*np.pi+2*np.pi/20,2*np.pi/20)
y = np.arange(0,2*np.pi+2*np.pi/20,2*np.pi/20)

X,Y = np.meshgrid(x,y)

u = np.sin(X)*np.cos(Y)
v = -np.cos(X)*np.sin(Y)

################################
# calculate magnitudes
################################
U = -1 - X**2 + Y
V = 1 + X - Y**2
my_color=u**2+v**2

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

ax1.set_title('Glyphs')
ax1.set_xlim(0, 6)
ax1.set_ylim(0, 6)
################################
# plot arrow glyphs
################################
ax1.quiver(x, y, u, v, scale=35)

ax2.set_title('Streamlines')
ax2.set_xlim(0, 6)
ax2.set_ylim(0, 6)
################################
# plot colored streamlines
################################
ax2.streamplot(X, Y, u, v,density=2,color=my_color,linewidth=2,cmap='viridis')

fig.show()
```
